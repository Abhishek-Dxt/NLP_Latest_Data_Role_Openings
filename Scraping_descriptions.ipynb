# Scraping Job Data from LinkedIn -

Job description for different roles can give important information about market trends, skill requirements, responsibilities, compensation, work location, company-wise opportunities etc.

### Part 1  - Getting the Job IDs

For any particular role, I will first scrape data using an API. The scraped data will include information ranging from job id, to job link, company name, job location etc.

```python
import requests

target_url='https://api.scrapingdog.com/linkedinjobs?api_key=<your_API_key>&field=Data%20Engineer&geoid=103644278&page=1'
# You will need an API Key of your own

resp = requests.get(target_url).json()

print(resp)
```

```python
job_id_list = []
for i in resp:
    if 'job_id' in i:
        job_id_list.append(i['job_id'])
```

```python
print(len(job_id_list))
```

### Part 2  - Getting Job Descriptions -

Having the job ids in hand, I have created a function that takes in a linkedin job id and returns the job description.

```python
import cfscrape
from bs4 import BeautifulSoup

def get_linkedin_job_description(job_id):
    job_url = f'https://www.linkedin.com/jobs/view/{job_id}/'
    scraper = cfscrape.create_scraper()
    response = scraper.get(job_url)

    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        description_div = soup.find('div', class_='description__text description__text--rich')
        job_description = description_div.get_text(strip=True) if description_div else "No job description found"
        return job_description
    else:
        print(response.status_code)
        return None
```

```python
for job_id in job_id_list[:5]:
    print(get_linkedin_job_description(job_id),'\n\n')
```
